```python
from database import DataBase
import pandas as pd
from yfinance import yahoo_finance
from features import Columns as Columns
```

```python
{key.value[0]:key.value[1] for key in Columns}


```

```python
# Main
db = DataBase("stock.db")

#fd = processor(db)


```

```python
def retrieve(symbol:str) -> None:
    if not db.check(table=symbol):
        db.create(table=symbol,columns={key.value[0]:key.value[1] for key in Columns})
    stock_scrape = yahoo_finance(db,db_symbol=symbol)
    stock_scrape(stock=symbol)
```

```python
for stk in ["SBUX","AAPL","MSFT","TSLA","CMCSA","AMZN"]:
    retrieve(symbol=stk)
```
